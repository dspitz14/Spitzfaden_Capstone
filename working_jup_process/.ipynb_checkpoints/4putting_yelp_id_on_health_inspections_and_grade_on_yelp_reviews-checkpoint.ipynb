```python
inspections_with_yelp = inspections.merge(mapper, left_on='permit_number', right_on='permit_number')
```

```python
inspections_with_yelp.sort_values('name_fuzz_ratio', ascending=False)
```

```python
reviews['yelp_review_date'] = pd.to_datetime(reviews['yelp_review_date'], errors='coerce')
```

```python
reviews.info()
```

```python
inspections_with_yelp['inspection_date'] = pd.to_datetime(inspections_with_yelp['inspection_date'], errors='coerce')
inspections_with_yelp['last_inspection']= pd.to_datetime(inspections_with_yelp['last_inspection'], errors='coerce')
```

```python
reviews.head()
```

```python
inspections_with_yelp.info()
```

```python
merged= reviews.merge(inspections_with_yelp, left_on='yelp_business_id', right_on='id')
```

```python
merged = merged[(merged['yelp_review_date']>merged['last_inspection']) & (merged['yelp_review_date']<=merged['inspection_date'])]
```

```python
merged = merged.drop_duplicates(subset= ['review_id', 'yelp_business_id'])
```

```python
merged.groupby(['yelp_business_id', 'inspection_date', 'last_inspection']).nunique().sort_values('review_id', ascending=False).head()
```
