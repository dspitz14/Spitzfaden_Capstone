```python
import pandas as pd
import numpy as np
import string
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
yelp_biz = pd.read_csv('/Users/dspitzfaden/Galvanize/capstone/data/all_nevada_biz.csv')
inspection_biz = pd.read_csv('/Users/dspitzfaden/Galvanize/capstone/data/restaurant_establishments.csv', sep=';')
```

```python
yelp_biz['yelp_address']= yelp_biz['address']
```

```python
stripped_yelp_address = []
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
non_dig= '''!()-[]{}abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ;:'"\,<>./?@#$%^&*_~'''

for i in range(len(yelp_biz)):
    name = str(yelp_biz.iloc[i]['yelp_address'])
    no_punct = ""
    for char in name:
        if char not in non_dig:
            no_punct = no_punct + char
#     lat = yelp_biz.iloc[i]['latitude']
#     print(name in inspection_biz['restaurant_name'])
#     print(long in inspection_biz['longitude'])
    stripped_yelp_address.append(no_punct.rstrip().lower())

    
stripped_inspections_address = []
for i in range(len(inspection_biz)):
    name = str(inspection_biz.iloc[i]['address'])
    no_punct = ""
    for char in name:
        if char not in non_dig:
            no_punct = no_punct + char
#     lat = yelp_biz.iloc[i]['latitude']
#     print(name in inspection_biz['restaurant_name'])
#     print(long in inspection_biz['longitude'])
    stripped_inspections_address.append(no_punct.rstrip().lower())
    


```

```python
yelp_biz['yelp_strip_address'] = stripped_yelp_address
```

```python
inspection_biz['inspect_strip_address'] = stripped_inspections_address
```

```python
merge_df=yelp_biz.merge(inspection_biz, left_on=['yelp_strip_address'], right_on = ['inspect_strip_address'])
merge_df = merge_df[pd.notnull(merge_df['yelp_strip_address'])]
```

```python
def getfuzz(col1, col2, n):
    fuzz_scores=[]
    for i in range(n):
        fuzz_scores.append(fuzz.ratio(str(col1.iloc[i]), str(col2.iloc[i])))
    return fuzz_scores
```

```python
name_fuzzies= getfuzz(merge_df['name'], merge_df['restaurant_name'], len(merge_df))
merge_df['name_fuzz_ratio']= name_fuzzies
```

```python
plt.xlabel("'Fuzz' Ratio Score", fontsize=15)
plt.ylabel('Count of Ratio Scores', fontsize=15)
plt.title('Business Name Match Histogram', fontsize=17)
merge_df['name_fuzz_ratio'].hist(bins=20)
```

```python
above_thresh = merge_df[(merge_df['name_fuzz_ratio']>= 45)]
above_thresh['name_fuzz_ratio'].hist(bins=40)

plt.xlabel("'Fuzz' Ratio Score", fontsize=15)
plt.ylabel('Count of Ratio Scores', fontsize=15)
plt.title('Business Name Match Histogram with Score above 45', fontsize=17)
```

```python
above_thresh = above_thresh[['id','facility_id','permit_number','name', 'restaurant_name', 'yelp_address', 'address_y', 'name_fuzz_ratio']].sort_values('name_fuzz_ratio', ascending=False)
```

```python
cleaned_above_thresh=above_thresh[['id','facility_id','permit_number','name', 'restaurant_name', 'yelp_address', 'address_y', 'name_fuzz_ratio']].sort_values('name_fuzz_ratio', ascending=False)
cleaned_above_thresh= cleaned_above_thresh[(cleaned_above_thresh['facility_id']!='FA0082600')|(cleaned_above_thresh['facility_id']!='FA0008423')|(cleaned_above_thresh['facility_id']!='FA0004910')|(cleaned_above_thresh['facility_id']!='FA0004969')|(cleaned_above_thresh['facility_id']!='FA0005087')|(cleaned_above_thresh['facility_id']!='FA0031638')|(cleaned_above_thresh['facility_id']!='FA0031679')|(cleaned_above_thresh['facility_id']!='FA0006336')|(cleaned_above_thresh['facility_id']!='FA0060862')|(cleaned_above_thresh['facility_id']!='FA0003331')|(cleaned_above_thresh['facility_id']!='FA0008867')|(cleaned_above_thresh['facility_id']!='FA0009125')]
cleaned_above_thresh=cleaned_above_thresh.dropna(how='any').sort_values('name_fuzz_ratio', ascending=False)
```

```python
cleaned_above_thresh.to_csv('/Users/dspitzfaden/Galvanize/capstone/data/new_yelp_gov_mapper.csv')
```
